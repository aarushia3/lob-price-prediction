```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../data/BenchmarkDatasets/Auction/1.Auction_Zscore/Auction_Zscore_Training/Train_Dst_Auction_ZScore_CF_1.txt")
print(data.shape)
```

```python
data[:5]
```

```python
X = data[:-5, :]
y = data[-5:, :]
```

```python
X.shape, y.shape
```

```python
# ‘1’ is for up-movement, ‘2’ is for stationary condition and ‘3’ is for down-movement.
unique, counts = np.unique(y, return_counts=True)

plt.bar(unique, counts)
plt.title("Label Distribution")
plt.show()
```

```python
plt.plot(X[0, :100])
plt.title("Sample Feature Over Time")
plt.show()
```

```python
plt.plot(y[:200])
plt.title("Label Sequence (First 200 steps)")
plt.show()
```

```python
df = pd.DataFrame(X)
corr = df.corr()

plt.imshow(corr, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.colorbar()
plt.show()
```
